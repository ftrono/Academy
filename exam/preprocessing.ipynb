```python
passenger_id = pd.read_csv(datadir + "/" + "passenger_id.csv")
passenger_id['name_orig'] = passenger_id['Surname'] + ", " + passenger_id['Name']
passenger_id
```

```python
board_registry = pd.read_csv(datadir + "/" + "board_registry.csv")
board_registry['name_orig'] = board_registry['Surname'] + ", " + board_registry['Name']
board_registry
```

```python
families = pd.read_csv(datadir + "/" + "families.csv")
families['name_orig'] = families['Surname'] + ", " + families['Name']
families
```

Import JSON:

```python
import json

tickets_sold = json.load(open("../out/Titanic/tickets_sold.json", 'r'))
tickets_sold[0:3]
```

```python
tickets_sold[-2:]
```

Reformat JSON:

```python
ls_tickets = []
ls_holders = []
ls_fares = []
ls_classes = []
ls_cabins = []

for item in tickets_sold:
    ls_tickets.append(item['Ticket'])
    ls_holders.append(item['Holder Name'])
    ls_fares.append(item['Fare'])
    ls_classes.append(item['Ticket Class'])
    ls_cabins.append(item['Cabin'])

tickets = pd.DataFrame({'Ticket': ls_tickets, 'name_orig': ls_holders, 'Fare': ls_fares, 'Ticket Class': ls_classes, 'Cabin': ls_cabins})
tickets
```
