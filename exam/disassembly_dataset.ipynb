## **Scomposizione Titanic Dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

class_names = {1: 'First Class', 2: 'Second Class', 3: 'Third Class'}
```

Import totale:

```python
full = pd.read_csv("../Dataset/Titanic/train.csv")
# test = pd.read_csv("../Dataset/Titanic/test.csv")
# full = pd.concat([train, test])
full
```

#### **1) Mapping Nomi - Passenger ID:**

```python
ids = full[['PassengerId', 'Name']]
ids.columns = ['Passenger Id', 'Name Orig']
ids['Surname'] = ids['Name Orig'].apply(lambda x: x.split(", ")[0])
ids['Name'] = ids['Name Orig'].apply(lambda x: x.split(", ")[1])
ids.sort_values(by='Passenger Id', inplace=True)
ids.reset_index(drop=True, inplace=True)
ids
```

```python
ids.drop(columns='Name Orig', inplace=True)
ids
```

```python
ids.to_csv("../out/Titanic/passenger_id.csv", index=False)
```

#### **2) Anagrafe passeggeri**:

```python
registry = full[['Name', 'Sex', 'Age', 'Pclass', 'Embarked']]
registry.sort_values(by='Name', inplace=True)
registry.rename(columns={'Name': 'Name Orig', 'Pclass': 'Pclass Orig'}, inplace=True)
registry.reset_index(drop=True, inplace=True)
registry
```

```python
registry['Pclass'] = [class_names[registry['Pclass Orig'].iloc[i]] for i in range(len(registry.index))]
registry['Surname'] = registry['Name Orig'].apply(lambda x: x.split(", ")[0])
registry['Name'] = registry['Name Orig'].apply(lambda x: x.split(", ")[1])
registry['Sex'] = np.where(registry['Sex'] == 'male', 'Male', 'Female')
registry
```

```python
registry_ok = pd.DataFrame({'Surname': registry['Surname'], 'Name': registry['Name'], 'Sex': registry['Sex'], 'Age': registry['Age'], 'Pclass': registry['Pclass'], 'Embarked': registry['Embarked']})
registry_ok
```

```python
registry_ok.to_csv("../out/Titanic/board_registry.csv", index=False)
```

#### **3) Biglietteria**

```python
tickets = full[['Name', 'Ticket', 'Fare', 'Cabin', 'Pclass']]
tickets.reset_index(drop=True, inplace=True)
tickets
```

```python
print(len(tickets['Ticket'].unique().tolist()))
```

Estrazione statistiche targets:

```python
targets = tickets[['Pclass', 'Fare']].groupby(by='Pclass', as_index=False).mean()
targets['Fare'] = targets['Fare'].round(4)
targets
```

```python
fares_map = {i: targets['Fare'].loc[i-1] for i in targets['Pclass']}
fares_map
```

Aggiunta persone esterne:

```python
real_extras = ['Hershey, Mr. Milton',
                'Morgan, Mr. J. Pierpont',
                'Marconi, Mr. Guglielmo',
                'Frick, Mr. Henry Clay',
                'Vanderbilt, Mr. Alfred Gwynne',
                'Dreiser, Mr. Theodore',
                'Mott, Mr. John',
                'Penney, Mr. James Cash',
                'Seiberling, Mr. Frank',
                'Dix, Mr. John Alden',
                'Adams, Mr. Henry']
```

```python
fakes = pd.read_csv("../Dataset/Titanic/nomi_fake.csv", sep=';')
fakes.head()
```

Add extras:

```python
import random
ls_extras = real_extras + fakes['Surname, Name'].tolist()
ls_names = []
ls_tickets = []
ls_pclasses = []
ls_fares = []

for extra in ls_extras:
    ls_names.append(extra)
    ls_tickets.append(f"ET {random.randint(10000000, 99999999)}")
    pclass = 1 if extra in real_extras else random.randint(1, 3)
    ls_pclasses.append(pclass)
    ls_fares.append(fares_map[pclass])

extras = pd.DataFrame({'Name': ls_names, 'Ticket': ls_tickets, 'Fare': ls_fares, 'Pclass': ls_pclasses})
extras.head(15)
```

```python
tickets = pd.concat([tickets, extras])
tickets
```

Prepare for export:

```python
tickets.sort_values(by='Ticket', inplace=True)
tickets['Ticket Class'] = [class_names[tickets['Pclass'].iloc[i]] for i in range(len(tickets.index))]
tickets.rename(columns={'Name': 'Name Orig'}, inplace=True)
tickets.reset_index(drop=True, inplace=True)
tickets
```

JSON version:

```python
from datetime import datetime, timedelta
import json, random

sale_start = datetime.strptime("1911-05-31 12:00:00", "%Y-%m-%d %H:%M:%S")
sale_end = datetime.strptime("1912-04-10 11:00:00", "%Y-%m-%d %H:%M:%S")
delta_minutes = int((sale_end - sale_start).total_seconds() / 60)
print(delta_minutes)
print(sale_start + timedelta(minutes=delta_minutes))
```

```python
tickets_sold = []
warned = []
bls = ['No', 'Yes']
#pets = ['Dog', 'Cat', 'Bird']

for t in tickets['Ticket'].unique().tolist():
    filt = tickets.loc[tickets['Ticket'] == t].reset_index()
    buyer = filt['Name Orig'].iloc[0]
    purchase_time = datetime.strftime(sale_start + timedelta(minutes=random.randint(0,delta_minutes)),"%Y-%m-%d %H:%M:%S")
    #print(purchase_time)
    for i in filt.index:
        buf = {}
        buf['Ticket'] = t
        buf['Purchase Time'] = purchase_time
        buf['Buyer Name'] = buyer
        buf['Holder Name'] = filt['Name Orig'].iloc[i]
        buf['Fare'] = filt['Fare'].iloc[i]
        buf['Ticket Class'] = filt['Ticket Class'].iloc[i]
        buf['Cabin'] = filt['Cabin'].iloc[i]
        has_pet = random.randint(0, 1)
        buf['Has Pet'] = bls[has_pet]
        tickets_sold.append(buf)

json.dump(tickets_sold, open("../out/Titanic/tickets_sold.json", 'w'))
```

#### **4) Famiglie**

```python
families = full[['Name', 'SibSp', 'Parch']].loc[(full['SibSp'] > 0) | (full['Parch'] > 0)]
families.sort_values(by='Name', inplace=True)
families.rename(columns={'Name': 'Name Orig', 'SibSp': 'Sib Sp', 'Parch': 'Par Ch'}, inplace=True)
families['Surname'] = families['Name Orig'].apply(lambda x: x.split(", ")[0])
families['Name'] = families['Name Orig'].apply(lambda x: x.split(", ")[1])
families.reset_index(drop=True, inplace=True)
families
```

```python
print(len(families['Surname'].unique().tolist()))
```

```python
families_ok = families[['Surname', 'Name', 'Sib Sp', 'Par Ch']]
families_ok
```

```python
families_ok.to_csv("../out/Titanic/families.csv", index=False)
```

#### **5) Sopravvissuti**

```python
survived = full[['Name', 'Sex', 'Age']].loc[full['Survived'] == 1]
survived.rename(columns={'Name': 'Name Orig'}, inplace=True)
survived['Surname'] = survived['Name Orig'].apply(lambda x: x.split(", ")[0])
survived['Name'] = survived['Name Orig'].apply(lambda x: x.split(", ")[1])
survived.reset_index(drop=True, inplace=True)
survived
```

```python
survived_ok = survived[['Surname', 'Name', 'Sex', 'Age']]
survived_ok['Rescue Ship'] = ['Carpathia' for i in survived_ok.index]
survived_ok['Arrival Pier'] = ['Pier 54' for i in survived_ok.index]
survived_ok['Arrival Harbour'] = ['New York' for i in survived_ok.index]
survived_ok['Arrival Time'] = [datetime.strptime("1912-04-18 21:30:00", "%Y-%m-%d %H:%M:%S") + timedelta(seconds=random.randint(0,342)) for i in survived_ok.index]
survived_ok
```

```python
survived_ok.to_csv("../out/Titanic/arrival_registry.csv", index=False)
```
