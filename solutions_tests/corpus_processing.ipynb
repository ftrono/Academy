### **Corpus processing**

```python
import statistics
```

#### Punto 1

```python
fcorpus = open("../Dataset/sample_corpus.txt", 'r')
corpus = []
stats = []

for line in fcorpus:
    sent = line.strip().split()
    corpus.append([word.lower() for word in sent])
    stats.append(len(sent))

fcorpus.close()

print("CORPUS:")
print(corpus)

print(f"Totale frasi: {len(stats)}")
print(f"Totale parole: {sum(stats)}")
print(f"Lunghezza media frasi: {round(statistics.mean(stats),2)} parole")
```

#### Punto 2

```python
freqs = {}

for sent in corpus:
    for word in sent:
        freqs[word] = freqs.get(word, 0) + 1

print(freqs)

vocabulary = list(freqs.keys())
print(vocabulary)
print(f"Lunghezza vocabulary: {len(vocabulary)} parole")
print(f"\nParole con massima e minima frequenza:")
print({k: v for k, v in freqs.items() if v == max(freqs.values())})
print({k: v for k, v in freqs.items() if v == min(freqs.values())})

print(f"\nVocabulary riordinato:")
print(sorted(vocabulary))
print(sorted(vocabulary, reverse=True))
```

#### Punto 4

```python
def word2index(corpus, vocabulary):
    vect_corpus = []
    for sent in corpus:
        buffer = []
        for word in sent:
            ind = vocabulary.index(word) if word in vocabulary else len(vocabulary)
            buffer.append(ind)
        vect_corpus.append(buffer)
    return vect_corpus
```

```python
def index2word(vect_corpus, vocabulary):
    conv_corpus = []
    for sent in vect_corpus:
        buffer = []
        for ind in sent:
            word = vocabulary[ind] if ind < len(vocabulary) else 'OOV'
            buffer.append(word)
        conv_corpus.append(buffer)
    return conv_corpus
```

```python
#Conversione:
vect_corpus = word2index(corpus, vocabulary)
print(vect_corpus)

conv_corpus = index2word(vect_corpus, vocabulary)
print(conv_corpus)
```

Export

```python
def export_corpus(corpus, outdir):
    outfile = open(outdir, 'w')
    for sent in corpus:
        outfile.write(f"{sent}\n")
    outfile.close()
    return 0
```

```python
export_corpus(vect_corpus, "../Out/corpus_vect.txt")
export_corpus(conv_corpus, "../Out/corpus_conv.txt")
```
