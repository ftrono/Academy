```python
from google.colab import drive, files

drive.mount("/content/drive/", force_remount=False)

#home directory:
%cd "/content/drive/My Drive/Lab_Python"

dataset = "./Dataset/sample_corpus.txt"
# dataset_mod = "./Out/sample_corpus_out.txt"

lista_frasi = []
lista_parole = []

sample_corpus = open(dataset, "r")
# sample_corpus_m = open(dataset_mod, "w")
i = 0
cont_parole = 0


for line in sample_corpus:
  line_corrected = line.lower().strip()
  line_corrected2 = line_corrected.split()
  lista_frasi.append(line_corrected2)
  cont_parole += len(line_corrected2)

numero_frasi = len(lista_frasi)

media = cont_parole / numero_frasi

print(numero_frasi)
print(cont_parole)
print(media)

```

```python
vocabulary = {}

for i in lista_frasi:
  for j in i:
    if j in vocabulary.keys():
      cont = vocabulary.get(j)
      cont += 1
      vocabulary[j] = cont
    else:
      vocabulary[j] = 1

# vocabulary

len_voc = len(vocabulary)
# len_voc

def CalcoliVoc(vocabulary_):
  freq = []

  for el in vocabulary_:
    freq.append(vocabulary_.get(el))

  for el in vocabulary_:
    if vocabulary_.get(el) == max(freq):
      print(f"la frequenza maggiore è {max(freq)} della parola {el}")
  for el in vocabulary_:
    if vocabulary_.get(el) == min(freq):
       print(f"la frequenza minore è {min(freq)} della parola {el}")

  print(f'la lunghezza del vocabolario è {len(vocabulary_)}')



lista_da_ord = []
l_ord1 = []
l_ord2 = []

for el in vocabulary:
  l_ord1.append(el)
  l_ord2.append(el)

l_ord1.sort()
l_ord2.sort(reverse=True)

CalcoliVoc(vocabulary)
```

```python
vocabulary_t = []
# list_idx2 = []
list_idx = []

for i in lista_frasi:
  for j in i:
    vocabulary_t.append(j)

def word2index():
  y = 0
  # newlist = []
  for ind in lista_frasi:
    for x in ind:
      if x in voc_nostop.keys():
        y = ind.index(x)
        list_idx.append(y)
      else:
        list_idx.append(-1)
    # list_idx.append(newlist)
  # return list_idx
word2index()

list_word = []

def index2word():
  for el in list_idx:
    if el == -1:
      list_word.append('OOV')
    else:
      y = vocabulary_t[el]
      list_word.append(y)

index2word()

func1 = "./Out/word2index_.txt"
func2 = "./Out/index2word.txt"

func1_ = open(func1, "w")
func2_ = open(func2, "w")

func1_.write(str(list_idx))
func2_.write(str(list_word))

func1_.close()
func2_.close()
sample_corpus.close()
```
