## **Punto 4**

```python
import copy
def word2index(main_list, stop_dict):
  code = main_list[:]
  lista_keys = list(stop_dict.keys())
  for i in range(len(main_list)):
    for j in range(len(main_list[i])):
      if main_list[i][j] in stop_dict:
        code[i][j] = lista_keys.index(code[i][j])
      else:
        code[i][j] = 1000
  return code
```

```python
numeri = word2index(main_list, stop_dict)
```

```python
numeri
```

```python
num = './Out/numeri1000.txt'
corpusnum = open(num, 'w')
for i in numeri:
  conv = map(str, i)
  string = ' '.join(conv)
  corpusnum.write(string + '\n') #scrivere ogni linea del file di output



```

```python
text = numeri.copy()

def index2word(text,stop_dict):
  lista_keys = list(stop_dict.keys())

  for i in range(len(text)):
    for j in range(len(text[i])):
      if int(text[i][j]) < len(lista_keys):
        text[i][j] = lista_keys[int(text[i][j])]
      else:
        text[i][j] = 'OOV'
      # try:
      #   text[i][j] = lista_keys[int(text[i][j])]
      # except:
      #   text[i][j] = 'OOV'
  return text
```

```python
text_res = index2word(text, stop_dict)
```
