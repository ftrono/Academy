```python
dataset = "./Dataset/sample_corpus.txt"
# dataset_modified =  "./Out/players_conv.csv"
corpus = []

corpus_file = open(dataset, 'r')
# players_csv = open(dataset_modified, 'w')
conta_frasi = 0
conta_parole = 0
lunghezza_totale_frasi = 0

for line in corpus_file:
  line_corrected = line.lower()
  line_corrected = line_corrected.strip()
  line_corrected = line_corrected.split()
  corpus.append(line_corrected)
  conta_frasi += 1
  conta_parole += len(line_corrected)

print(corpus)
print(f'Il numero di frasi è {conta_frasi}')
print(f'Il numero di parole è {conta_parole}')
media = conta_parole/conta_frasi
print(f'La lunghezza media delle frasi è {round(media,2)}')

corpus_file.close()
```

Esercizio 2

```python
dic = {}
# Conto le parole
for frase in corpus:
  for parola in frase:
    if parola not in dic.keys():
      dic[parola] = 1
    else:
      dic[parola] += 1

print(dic)
lunghezza = len(dic)
print(f'La lunghezza del vocabolario è {lunghezza}')

# frequenza minima
frequenza_min = 0

for v in dic.values():
  if v == 1:
    frequenza_min += 1
print(f"Numero parole di frequenza minima: {frequenza_min}")

for key, value in dic.items():
  if value == 1:
    print(key)

# frequenza massima
v_max = max(list(dic.values()))
frequenza_massima = 0
for v in dic.values():
  if v == v_max:
    frequenza_massima += 1
print(f"Numero parole di frequenza minima: {frequenza_massima}")

for key, value in dic.items():
  if value == v_max:
    print(key)
```

```python
# parole vocabolario in ordine crescente e decrescente
lista_parole = []
for k in dic.keys():
  lista_parole.append(k)

lista_ordine_alfabetico = sorted(lista_parole)
lista_ordine_decrescente = sorted(lista_parole, reverse = True)

print(lista_ordine_alfabetico)
print(lista_ordine_decrescente)
```

```python
def word2index(dic_stop_words, corpus):
  ls_stop_words = list(dic_stop_words.keys())
  ls_num = []
  for lista in corpus:
    vettore = []
    for element in lista:
      if element in ls_stop_words:
        element = ls_stop_words.index(element)
        vettore.append(element)
      else:
        element = -1
        vettore.append(element)
    ls_num.append(vettore)
  return ls_num

ls_num = word2index(dic_stop_words, corpus)
ls_num
```
