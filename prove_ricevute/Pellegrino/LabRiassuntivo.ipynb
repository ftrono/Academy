```python
sample_words = open("./Dataset/sample_corpus.txt", "r")
sample_list = [sentence.lower().strip().split() for sentence in sample_words]
sample_words.close()

cnt_sentence = len(sample_list)
print(cnt_sentence)

cnt_words = 0
for sentence in sample_list:
  for word in sentence:
    cnt_words += 1
print(f"cnt_words {cnt_words}")

sentence_lengths = [len(sentence) for sentence in sample_list]
avg_sentence_lengths = sum(sentence_lengths) / cnt_sentence
print(f"avg_sentence_lengths = {avg_sentence_lengths}")

vocabulary = {}
for sentence in sample_list:
  for word in sentence:
    try:
      vocabulary[word] += 1
    except KeyError:
      vocabulary[word] = 1

cnt_vocabulary = len(vocabulary)
print(f"cnt_vocabulary = {cnt_vocabulary}")

# NOTE: this is very ugly, find something better
def find_min(vocabulary: dict):
  min_freq = 1000000
  min_freq_words = []

  for key, value in vocabulary.items():
    if value < min_freq:
      min_freq = value
      min_freq_words.clear()
      min_freq_words.append(key)
    elif value == min_freq:
      min_freq_words.append(key)
  return min_freq_words, min_freq

def find_max(vocabulary: dict):
  max_freq = 0
  max_freq_words = []

  for key, value in vocabulary.items():
    if value > max_freq:
      max_freq = value
      max_freq_words.clear()
      max_freq_words.append(key)
    elif value == max_freq:
      max_freq_words.append(key)
  return max_freq_words, max_freq

min_freq_words, min_freq = find_min(vocabulary)
max_freq_words, max_freq = find_max(vocabulary)

print(f"min_freq_words = {min_freq_words}, min_freq = {min_freq}")
print(f"max_freq_words = {max_freq_words}, max_freq = {max_freq}")

# order words in decreasing and increasing order (alphabeticallyl)
words_increasing_order = list(vocabulary.keys())
words_increasing_order = sorted(words_increasing_order)
words_decreasing_order = words_increasing_order[::-1]
print(words_increasing_order)
print(words_decreasing_order)

# rm words that appears < 2 times
vocabulary_stripped = {key: val for key, val in vocabulary.items() if val >= 2}
print(len(vocabulary_stripped))
min_freq_words, min_freq = find_min(vocabulary_stripped)
max_freq_words, max_freq = find_max(vocabulary_stripped)
print("First strip")
print(f"min_freq_words = {min_freq_words}, min_freq = {min_freq}")
print(f"max_freq_words = {max_freq_words}, max_freq = {max_freq}")

# rm words that appears > 100
vocabulary_stripped = {key: val for key, val in vocabulary_stripped.items() if val <= 100}
print(len(vocabulary_stripped))
min_freq_words, min_freq = find_min(vocabulary_stripped)
max_freq_words, max_freq = find_max(vocabulary_stripped)
print("Second strip")
print(f"min_freq_words = {min_freq_words}, min_freq = {min_freq}")
print(f"max_freq_words = {max_freq_words}, max_freq = {max_freq}")

# rm words that appears in stop_words.txt file
stop_words = open("./Dataset/stop_words.txt", "r")
stop_list = [word.strip().lower() for word in stop_words]
stop_words.close()

vocabulary_stripped = {key: val for key, val in vocabulary_stripped.items() if key not in stop_list}
print(len(vocabulary_stripped))
min_freq_words, min_freq = find_min(vocabulary_stripped)
max_freq_words, max_freq = find_max(vocabulary_stripped)
print("Third strip")
print(f"min_freq_words = {min_freq_words}, min_freq = {min_freq}")
print(f"max_freq_words = {max_freq_words}, max_freq = {max_freq}")

import copy

def word2index(corpus_list: list, vocabulary_stripped: dict) -> list:
  d = {}
  for idx, key in enumerate(vocabulary_stripped):
    d[key] = idx

  idx_list = copy.deepcopy(corpus_list)
  for i, sentence in enumerate(corpus_list):
    for j, word in enumerate(sentence):
      if word in d:
        idx_list[i][j] = d[word]
      else:
        idx_list[i][j] = -1
  return idx_list

def index2word(token_list: list, vocabulary_stripped: dict) -> list:
  d = {}
  for idx, key in enumerate(vocabulary_stripped):
    d[idx] = key

  corpus_list = copy.deepcopy(token_list)
  for i, sentence_tok in enumerate(token_list):
    for j, tok in enumerate(sentence_tok):
      if tok == -1:
        corpus_list[i][j] = 'OOV'
      else:
        corpus_list[i][j] = d[tok]
  return corpus_list

token_list = word2index(sample_list, vocabulary_stripped)
reconverted_corpus_list = index2word(token_list, vocabulary_stripped)

# export token_list and reconverted_corpus list
token_file = open("./Out/token.txt", "w")
corpus_file = open("./Out/corpus.txt", "w")

str_token_list = "".join(str(token_list))
str_reconverted_corpus_list = "".join(str(reconverted_corpus_list))
token_file.write(str_token_list)
corpus_file.write(str_reconverted_corpus_list)

token_file.close()
corpus_file.close()
```
