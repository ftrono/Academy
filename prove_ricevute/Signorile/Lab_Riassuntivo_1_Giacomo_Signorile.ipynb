```python
dataset = "./Dataset/sample_corpus.txt"
#dataset_modified = "./Out/players_conv.txt"

lista_parole = []
lista = []

myfile = open(dataset, "r")
#outfile = open(dataset_modified, "w")

numero_frasi = 0
numero_parole = 0
i=0

for line in myfile:
  line_corrected = line.strip()
  line_lower = line_corrected.lower()
  line_list = line_lower.split()
  lista_parole.append(line_list)
  i+=1
  #print(len(line_list))
  for el in line_list:
    numero_parole += 1

numero_frasi = i
media_parole = numero_parole/numero_frasi

media_parole = round(media_parole, 3)

myfile.close()
#outfile.close()

print(f"Il numero di frasi è {numero_frasi}, e la lunghezza totale di parole nel dataset è {numero_parole}, la media di parole nel dataset è {media_parole}")
```
