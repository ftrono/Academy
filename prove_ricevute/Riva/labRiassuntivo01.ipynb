## 1. Statistiche sul corpus

```python
dataAddress = './Dataset/sample_corpus.txt'
dataFile = open(dataAddress, 'r')

########################################################################
# Lettura e sistemazione in lista di liste del corpus


corpus = []
phraseCounter = 0
phraseWordLen = []
for phrase in dataFile:
    phraseClean = phrase.lower()
    phraseStrip = phraseClean.strip()
    phraseSplit = phraseStrip.split()
    corpus.append(phraseSplit)
    # print(phraseCounter, len(phraseSplit))      #print di controllo
    phraseWordLen.append(len(phraseSplit))
    phraseCounter += 1



dataFile.close()
########################################################################
# corpus


phraseWordLenMean = int(round(stat.mean(phraseWordLen), 0))
wordSum = sum(phraseWordLen)

print(f'-Il numero di frasi nel corpus è {phraseCounter}.\n-Le parole sono {wordSum}.\n-Ci sono in media {phraseWordLenMean} parole in ogni frase.')
```

## 2. Realizzo un lessico

```python
vocabulary = {}
for phrase in corpus:
    for word in phrase:
        try:
            vocabulary[word] += 1
        except:
            vocabulary[word] = 1

# vocabulary

freqWordMax = [key for key, value in vocabulary.items() if value == max(vocabulary.values())]
freqWordMin = [key for key, value in vocabulary.items() if value == 1]

vocabularyLen = len(vocabulary)
print(f"""-Ci sono {vocabularyLen} voci nel vocabolario.\n-La voce con massima frequenza è "{freqWordMax[0]}"\n""")
print(f"""-Ci sono {len(freqWordMin)} voci con frequenza 1, e sono: """)
# for minimaFreqWord in freqWordMin:
#     print(f"""'{minimaFreqWord}': 1,""")


# due modi per fare la stessa cosa


for key, value in vocabulary.items():
    if value == 1:
        print(key)
```

## 3. Pulizia del `Vocabulary`

```python
################################################
#   Rimuovo le voci che appaiono meno di 2 volte
vocabularyMin = {}
for word in ascendingVoc:
    value = vocabulary.get(word)
    if value > 1:
        vocabularyMin[word] = value

vocMinFreqWordMax = [key for key, value in vocabularyMin.items() if value == max(vocabularyMin.values())]
vocMinFreqWordMin = [key for key, value in vocabularyMin.items() if value == min(vocabularyMin.values())]

print(f"""Il numero di voci tolte quelle con frequenza minore di 2 è {len(vocabularyMin)},\nparole con frequenza minima {len(vocMinFreqWordMin)} e frequenza massima {len(vocMinFreqWordMax)}.\n""")
print(f'Parole con frequenza minima {vocMinFreqWordMin}')
print(f'Parole con frequenza massima {vocMinFreqWordMax}')
print(f'Il vocabolario con le rispettive frequenze: {vocabularyMin}')
```

```python
#################################################
#   Rimuovo le voci che appaiono più di 100 volte

vocabularyMax = {}
for key, value in vocabularyMin.items():
    if value < 101:
        vocabularyMax[key] = value

vocMaxFreqWordMax = [key for key, value in vocabularyMax.items() if value == max(vocabularyMax.values())]
vocMaxFreqWordMin = [key for key, value in vocabularyMax.items() if value == min(vocabularyMax.values())]
print(f"""mentre il numero di voci tolte anche quelle con frequenza maggiore di 100 è {len(vocabularyMax)}\nparole con frequenza minima {len(vocMaxFreqWordMin)} e frequenza massima {len(vocMaxFreqWordMax)}.\n""")
print(f'Parole con frequenza minima {vocMaxFreqWordMin}')
print(f'Parole con frequenza massima {vocMaxFreqWordMax}')
print(f'Il vocabolario con le rispettive frequenze: {vocabularyMax}')
```

```python
###########################################
#   Rimuovo le voci dal file stop_words.txt
#   Read stop words data file
dataAddress = './Dataset/stop_words.txt'
stopWordsFile = open(dataAddress, 'r')

stopWords = []

for word in stopWordsFile:
    word = word.lower().strip()
    stopWords.append(word)
stopWordsFile.close()

vocabularyCensrd = {}

for key, value in vocabularyMax.items():
    if key not in stopWords:
        vocabularyCensrd[key] = value

vocCensFreqWordMax = [key for key, value in vocabularyCensrd.items() if value == max(vocabularyCensrd.values())]
vocCensFreqWordMin = [key for key, value in vocabularyCensrd.items() if value == min(vocabularyCensrd.values())]
print(f"""Il numero di voci del vocabolario senza le stop_words.txt è {len(vocabularyCensrd)},\nparole con frequenza minima {len(vocCensFreqWordMin)} e frequenza massima {len(vocCensFreqWordMax)}.\n""")
print(f'Parole con frequenza minima {vocCensFreqWordMin}')
print(f'Parole con frequenza massima {vocCensFreqWordMax}')
print(f'Il vocabolario con le rispettive frequenze: {vocabularyCensrd}')
```

## 4.  Vettorizzazione del dizionario

```python
def word2index(corpus, vocabulary):
    words = []
    for wordKey in vocabulary.keys():
        words.append(wordKey)
    OOV = len(words) + 1

    transCorpus = []
    for phrase in corpus:
        phraseIndexed = []
        for word in phrase:
            if word in words:
                phraseIndexed.append(words.index(word))
            elif word not in words:
                # print(phraseCounter, wordCounter)
                phraseIndexed.append(OOV)
        transCorpus.append(phraseIndexed)
    return transCorpus

corpusIndexed = word2index(corpus, vocabularyCensrd)
```

```python
def index2word(indexedCorp, vocabulary):
    words = []
    for wordKey in vocabulary.keys():
        words.append(wordKey)
    OOV = len(words) + 1

    transCorpus = []
    for indexPhrase in indexedCorp:
        phraseUnindexed = []
        for index in indexPhrase:
            if index < len(words):
                phraseUnindexed.append(words[index])
            elif index == OOV:
                phraseUnindexed.append('OOV')
        transCorpus.append(phraseUnindexed)
    return transCorpus

corpusUnindexed = index2word(corpusIndexed, vocabularyCensrd)
```

## Esporto in formato `.txt` i risultati delle funzioni di traduzione

```python
outFile = open('./Out/indexed.txt', 'w')

for phraseIndexed in corpusIndexed:
    phrase = []
    for index in phraseIndexed:
        phrase.append(str(index))
    Frase = ' '.join(phrase)
    outFile.write(Frase + '\n')

outFile.close()
# files.download()#
```

```python
outFile = open('./Out/unindexed.txt', 'w')

for phraseUnindexed in corpusUnindexed:
    phrase = []
    for word in phraseUnindexed:
        phrase.append(str(word))
    Frase = ' '.join(phrase)
    outFile.write(Frase + '\n')

outFile.close()
```
