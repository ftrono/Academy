```python
dataset = "./Dataset/sample_corpus.txt"
```

```python
# Apro il file di dati in modalità di lettura
corpus_file = open(dataset, 'r')

```

```python
# creo lista vuota per memorizzare i dati

corpus = []
```

```python
for line in corpus_file:

    line_corrected = line.strip()

    line_list_min = line_corrected.lower()

    line_list = line_list_min.split()

    corpus.append(line_list)

    print(corpus)


# Chiudi il file di dati
corpus_file.close()
```

```python
print(corpus)
```

```python
# numero di frasi presenti
num_frasi = len(corpus)
print(num_frasi)
```

```python
# Conteggio parole
def contatore(lista):
  count = 0
  for el in lista:
    for word in el:
      count += 1
  return count

contatore(corpus)
num_parole = contatore(corpus)
```

```python
print(num_parole)
```

```python
media_frase = num_parole/num_frasi
print(media_frase)
```

```python
# Dizionario vuoto
d = {}

for lista in corpus:
  for word in lista:
    if word not in d.keys():
      d[word] = 1
    else:
      cont = d[word]
      cont += 1
      d[word] = cont

```

```python
print(d)
```

```python
lunghezza_d = len(d)
print(lunghezza_d)
```

```python
freq = []

for el in d:
  freq.append(d.get(el))

for el in d:
  if d.get(el) == min(freq):
    print(f"{el} {d.get(el)}")

for el in d:
  if d.get(el) == max(freq):
    print(f"{el} {d.get(el)}")
```

```python
#Ordine alfabetico crescente
for key in sorted(d.keys()):
  print(key, d[key])
```

```python
#Ordine alfabetico decrescente
for key in sorted(d.keys(), reverse = True):
  print(key, d[key])
```
