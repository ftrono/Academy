```python
dataset = "./Dataset/sample_corpus.txt"
dataset_mod = "./Out/sample_corpus_conv.txt"
```

```python
corpo_file = open(dataset, "r")
```

```python
corpo = []
for line in corpo_file:
  line_corrected = line.strip()
  line_corrected = line_corrected.lower()
  line_list = line_corrected.split()
  corpo.append(line_list)
```

```python
somma_parole = 0
for i in range(len(corpo)):
  somma_parole += len(corpo[i])

somma_parole
```

```python
vocabolario = {}
for i in range(len(corpo)):
  for j in range(len(corpo[i])):
   if vocabolario.get(corpo[i][j],-1) == -1:
      vocabolario[corpo[i][j]] = 1
   else:
    vocabolario[corpo[i][j]] += 1

vocabolario
```

```python
def word2index(stringa):
  parole = list(vocabolario.keys)
  codice = []
  stringa_c = stringa.strip()
  stringa_c = stringa.lower()
  stringa_l = stringa.split()

  for el in stringa_l:
    if el in parole:
      codice_p = parole.index(el)
    else:
      codice_p = -1
    codice.append(codice_p)
  return codice

word2index("the fox zombie yourself")
```
