# ESERCIZIO 1

```python
datafile = open(dataset, "r")
outfile = open(dataset_modified, "w")

corpus = []

#converto il corpus in lista
for line in datafile:
  line_corrected = line.strip()
  line_corrected = line_corrected.lower()
  line_list = line_corrected.split()
  corpus.append(line_list)

print(corpus)

#calcolo il numero delle frasi del dataset
num_frasi = len(corpus)
print (f"Le frasi sono {num_frasi}")

#calcolo il numero di parole totali del dataset
num_parole = 0
for i in range(len(corpus)):
  num_parole += len(corpus[i])

print (f"Le parole totali del dataset sono {num_parole}")

#calcolo la lunghezza media di ogni frase
sum = 0
for i in range(len(corpus)):
  sum += len(corpus[i])

media = sum/num_frasi
print (f"La lunghezza media delle frasi è {media:,.2f}")


datafile.close()
outfile.close()
```

```python

#creo un vocabulary con tutte le parole salvate una sola volta
vocabulary = []

for i in range(len(corpus)):
  for j in range(len(corpus[i])):
    if (corpus[i][j]) not in vocabulary:
      vocabulary.append(corpus[i][j])

print(vocabulary)


#conto la frequenza delle parole nel vocabulary
dizionario = {}
for i in range(len(corpus)):
  for word in corpus[i]:
    if word in dizionario:
      dizionario[word] += 1
    else:
      dizionario[word] = 1

print(dizionario)


# calcolo la lunghezza del vocabulary
print(f"\nIl vocabulary possiede {len(vocabulary)} parole")


#trovo la parola con frequenza minima e quella con frequenza massima

print(f"\nParole con massima e minima frequenza:")

print({k: v for k, v in dizionario.items() if dizionario[k] == max(dizionario.values())})

print({k: v for k, v in dizionario.items() if dizionario[k] == min(dizionario.values())})


#ordino le parole del vocabulary in ordine alfabetico crescente e decrescente
voc_cresc = sorted(vocabulary)
voc_dec = sorted(voc_cresc, reverse=True)

print("\nDizionario in ordine crescente:")
print(voc_cresc)
print("\nDizionario in ordine decrescente:")
print(voc_dec)

```
